# EXOPLANETARY DATA ANAYLSIS
The planets that are outside of our solar system are called exoplanets. The first ever exoplanet detected in 1972. here in the data that I used  has various properties of exoplanets detected and i used that to draw a conclusion from it.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import opendatasets as od
dataset_url = 'https://www.kaggle.com/shivamb/all-exoplanets-dataset'
od.download('https://www.kaggle.com/shivamb/all-exoplanets-dataset')
```

```python
import pandas as pd
raw_df=pd.read_csv('all-exoplanets-dataset/all_exoplanets_2021.csv')
raw_df
```

```python
raw_df.shape
```

```python
raw_df.info()
```

```python
selected_columns=['Planet Name', 
                  'Planet Host', 
                  'Num Stars',                       
                  'Num Planets',
                  'Discovery Method',                
                  'Discovery Year',                   
                  'Discovery Facility',             
                  'Orbital Period Days',
                  'Orbit Semi-Major Axis',          
                  'Mass',                                                             
                  'Equilibrium Temperature',        
                  'Spectral Type',                   
                  'Stellar Effective Temperature', 
                  'Stellar Radius', 
                  'Stellar Mass',
                  'Stellar Metallicity Ratio', 
                  'Distance',
                  'Gaia Magnitude']
```

```python
len(selected_columns)
```

```python
exoplanets_df=raw_df[selected_columns].copy()
exoplanets_df
```

```python
exoplanets_df.columns
```

```python
exoplanets_df['Discovery Year'] = exoplanets_df['Discovery Year'].apply(str)
```

```python
exoplanets_df.describe()
```

#### Detection  methods


```python
discometh_df=exoplanets_df['Discovery Method'].value_counts()
discometh_df
```

```python
plt.figure(figsize=(18,6))
plt.title('No of planets discovered by various detection methods')
sns.barplot(x=discometh_df.index, y=discometh_df);
```

The  exoplanets are usually detected by the above methods. but the transit method is more popuplar but still had its own disadvantages. transit method involves use of light curve and to observe it 

#### Discovered year

```python
discoyear_df=exoplanets_df.groupby('Discovery Year')[['Planet Name']].count()
discoyear_df
```

```python
plt.figure(figsize=(12,6))
plt.ylabel('No of planets')
plt.xlabel("years")
plt.title('No of planets discovered over years')
plt.plot(discoyear_df.index,discoyear_df['Planet Name'])
```

As we know that our space missions has incresed in capacity and accuracy we observe that there is a trend of increase in the exoplanet detection and confirmation. the launch of kepler mission helped us to acheive this feat. however in 2016 neary 1526 planets are confirmed and published 

##### Discovery facility

```python
discofacility_df=exoplanets_df['Discovery Facility'].value_counts()

disco_df=discofacility_df.head(10)
discofacility_df
```

```python
plt.figure(figsize=(18,6))
plt.title('No of planets discovered at various discover facilities')
sns.barplot(x=disco_df.index, y=disco_df);
```

There are many observatories established  around  the world . some of them usually in space like kepler and others are ground based establishment. The kepler has detected many planets than any other observatories 

#### Number of planets in the host system 

```python
sypnum_df=exoplanets_df['Num Planets'].value_counts()
sypnum_df
```

```python
plt.figure(figsize=(18,6))
plt.xlabel('no of planets in respective solar system')
plt.title('No of stars with with respective number of planets')
sns.barplot(x=sypnum_df.index, y=sypnum_df);
```

The solar system usaully consits of planets  revolving around the star. From the above data we can see that star with single planet revolving around it are more in number. Exoplanets are deteted are usually gravitionally bounded to a star or sometimes it can be a rogue planet.

#### Number of stars in system ( whether the stellar system is binary trinary or quaternary stellar system) 

```python
starnum_df=exoplanets_df['Num Stars'].value_counts()
starnum_df
```

```python
plt.figure(figsize=(12,6))
plt.xlabel('no of stars in respective solar system')
plt.title('No of stars in steller system')
sns.barplot(x=starnum_df.index, y=starnum_df);
```

In many cases stars can bound to each other gravitionally and revolve around a centre point. The stellar system with two stars is call binary star system with three terinary starsystem and four quatarnery star system. the single star system are more in number in the given data. 

#### Spectral type of stars

```python
spectral_df=exoplanets_df['Spectral Type'].value_counts()
spectral1_df=spectral_df.head(10)
spectral_df
```

```python
plt.figure(figsize=(18,6))
plt.xlabel('spectral type')
plt.title('No of stars with spectral types')
sns.barplot(x=spectral1_df.index, y=spectral1_df);
```

Stars are divided into many types in which spectral classification is one. the spectral classification follows Morgan–Keenan (MK) system which  uses the letters O, B, A, F, G, K, and M, a sequence from the hottest (O type) to the coolest (M type). Each letter class is then subdivided using a numeric digit with 0 being hottest and 9 being coolest. The most type of are G0V type.
